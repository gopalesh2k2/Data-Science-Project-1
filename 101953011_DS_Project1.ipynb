```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
data = pd.read_excel('DATASET.xlsx')
data.head()
```

```python
data = data.drop([910, 911, 912, 913, 914, 915, 916, 917, 918])
data.info()
```

```python
data.fillna(0, inplace=True)
data.info()
```

```python
j = 0
for i in range(100):
  data.iloc[j:j + 10, 0] = data.iloc[j, 0]
  j += 10

d_test = data.loc[data['year'] == 10]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
data.columns = data.columns.str.replace('Unnamed: 0', 'Group')
d_test.columns = d_test.columns.str.replace('Unnamed: 0', 'Group')
model = RandomForestRegressor(bootstrap=False, max_leaf_nodes=20,min_samples_leaf=2,min_weight_fraction_leaf=0.1,oob_score=False, verbose=0,warm_start=False)
```

```python
data_name=[]
data_name.append('9')
data_name.append('10')
data_name.append('11')
data_name.append('12')
data_name.append('13')
n = 0
ind = 0
root_mean = []
data_pre = {'9': [], '10': [], '11': [], '12': [], '13': []}
list=[10, 11, 12, 13, 14]
for j in list:
  a = 0
  co = 0
  for i in range(0, 1000, 10):
    x_testing = data.iloc[i+9, 1:10].to_numpy()
    y_testing = data.iloc[i+9, j]
    x_training = data.iloc[i:i+9, 1:10]
    y_training = data.iloc[i:i+9, j]
    model.fit(x_training, y_training)
    y_pre = model.predict(x_testing.reshape(1,-1))[0]
    data_pre[data_name[ind]].append(y_pre)
    a += (y_testing - y_pre)*(y_testing - y_pre)
    co += 1
    n += 1
  root_mean.append(a)
  ind += 1
```

```python
data_predict = pd.DataFrame.from_dict(data_pre)
data_predict
```

```python
print((sum(root_mean)/n)**0.5)
```
